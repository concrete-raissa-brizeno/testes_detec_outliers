```python
import numpy as np
import pandas as pd
from openpyxl import Workbook
from openpyxl.compat import range
from openpyxl.utils import get_column_letter
from openpyxl import load_workbook
```

```python
workBook = load_workbook(filename = '../DadosFonte/Dados condensados.xlsm')
sheet = workBook['Sheet1']
#print(sheet['CR17'].value)
```

```python
# Funcao para realizar os testes
def testes_outliers():
    mad_test()

# Funcao de test MAD
def mad_test(data):
    mediana = np.median(data)
    desvios_absolutos = np.abs(data - mediana)
    mad = np.median(desvios_absolutos)
    mad_e = 1.4826 * mad
    return data[((desvios_absolutos / mad_e) > 2)]
```

```python
def percorre_planilha(sheet, start_row, start_col):
    count = 0
    for col in range(start_col, sheet.max_column + 1):
        values = []
        for row in range(start_row, sheet.max_row + 1): 
            if(not(sheet.cell(row, col).value is None)):
                values.append(sheet.cell(row, col).value)
            else:
                # Realizar Testes em values
                #testes_outliers()
                print('\n\n')
                print(sheet.cell(row, col))
                print(values)
                count += 1
                print('\n\n')
                break
    print(count)
```
